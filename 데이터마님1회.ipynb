```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
df.head()

#1. 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 인원을 가진 나이대는? (0~9 : 0 , 10~19 : 10)
df['age_2'] = df['age'] // 10 * 10
df.head()
df['age_2'].value_counts().index[0]
```

```python
#2. 마케팅 응답 고객들의 나이를 10살 단위로 변환 했을 때, 가장 많은 나이대 구간의 인원은 몇명인가?
df2 = df['age_2'].value_counts().reset_index()
df2.iloc[0,1]
```

```python
#나이가 25살 이상 29살 미만인 응답 고객들중 housing컬럼의 값이 yes인 고객의 수는?
df.head()
cond1 = df['age'] >= 25
cond2 = df['age'] < 29
cond3 = df['housing'] == 'yes'

len(df[cond1 & cond2 & cond3])
```

```python
#4. numeric한 값을 가지지 않은 컬럼들중 unique한 값을 가장 많이 가지는 컬럼은?
df.head()
df.info()
df.describe(include = "O")
# df.describe()
```

```python
#balance 컬럼값들의 평균값 이상을 가지는 데이터를 ID값을 기준으로 내림차순 정렬했을때 상위 100개 데이터의 balance값의 평균은?
avg = df['balance'].mean()
avg
cond1 = df['balance'] >= avg
df2 = df[cond1].sort_values('ID', ascending = False)
df2 = df2.iloc[0:100, :]
print(df2['balance'].mean())
```

```python
#데이터의 job이 unknown 상태인 고객들의 age 컬럼 값의 정규성을 검정하고자 한다. 샤피로 검정의 p-value값을 구하여라
cond = df['job'] == "unknown"
df2 = df[cond]
from scipy.stats import shapiro
model = shapiro(df2['age'])
print(model)
```

```python
#각 job에 따라 divorced/married 인원의 비율을 확인 했을 때 그 값이 가장 높은 값은?
df.head()
df2 = df.groupby(['job', 'marital']).size().reset_index()
df3 = df2.pivot_table(index = "job", columns = "marital")[0]
df3 = df3.fillna(0)
df3['rate'] = df3['divorced'] / df3['married']
df3.sort_values('rate', ascending = False)
```

```python
import pandas as pd
train= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/train.csv')
test= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/test.csv')
submission= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bank/submission.csv')

display(train.head())
display(test.head())
display(submission.head())

train.info()

#test.describe(include = "O") job의 가짓수가 너무 많다 = 인코딩은 라벨인코딩으로 진행한다
print(train.shape)
test_id = test.pop("ID")
train = train.drop("ID", axis = 1)
print(train.shape, test_id.shape)

train.head()
cols= ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

#라벨 인코딩 : job이 너무 많아서 어쩔 수 없음
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in cols:
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])

train.head()


target = train.pop("y")

#train_test_split
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)
# print(x_tr.shape, x_val.shape, y_tr.shape, y_val.shape) #(10296, 15) (2574, 15) (10296,) (2574,)

#반응 여부 예측 : 분류 predict
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state = 0)
rfc.fit(x_tr, y_tr)
pred = rfc.predict_proba(x_val)[:,1]
pred

from sklearn.metrics import roc_auc_score
ras = roc_auc_score(y_val, pred)
print(ras) #0.7823523728774455

pred = rfc.predict_proba(test)[:, 1]
pred

submission = pd.DataFrame({'ID' : test_id, 'pred' : pred}).to_csv("00000000.csv", index = False)
pd.read_csv("00000000.csv")
```
